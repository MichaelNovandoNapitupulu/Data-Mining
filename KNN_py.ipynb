```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Data Intelligent Irrigation System.csv')
```

```python
dataset.head()
```

```python
x = dataset.iloc[:, [2,3]].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test  = sc.fit_transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Inisialisasi KNeighborsClassifier dengan metric 'minkowski'
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)

# Melatih model dengan data training
classifier.fit(x_train, y_train)
```

```python
y_pread = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pread)
print(cm)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Mengatur data untuk visualisasi
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(
    np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
    np.arange(start=x_set[:, 1].min() - 1, stop=x_set[:, 1].max() + 1, step=0.01)
)

# Plotting the decision boundary
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

# Plotting the data points
colors = ListedColormap(('red', 'green'))
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c=colors(i), label=j)

# Menambahkan judul dan label
plt.title('Klasifikasi Data dengan KNN (Data Training)')
plt.xlabel('Umur')
plt.ylabel('Estimasi Gaji')
plt.legend()
plt.show()
```
